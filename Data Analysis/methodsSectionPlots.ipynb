```python
# Library Imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Loading data and sample selection
dataFrame1 = pd.read_csv("REFL0100N1504galaxyData.txt", names=["x", "y", "z", "gn", "sgn", "mass", "vx", "vy", "vz", "sfr", "id", "r", "r2", "r3"])
dataFrame1 = dataFrame1[dataFrame1["mass"] > 3e9]

dataFrame2 = pd.read_csv("NoAGNL0050N0752galaxyData.txt", names=["x", "y", "z", "gn", "sgn", "mass", "vx", "vy", "vz", "sfr", "id", "r", "r2", "r3"])
dataFrame2 = dataFrame2[dataFrame2["mass"] > 3e9]

# Extracting columns
mass1 = dataFrame1["mass"].to_numpy()
sfr1 = dataFrame1["sfr"].to_numpy()

mass2 = dataFrame2["mass"].to_numpy()
sfr2 = dataFrame2["sfr"].to_numpy()

# Logging data
logMass1 = np.log10(mass1)
logSfr1 = np.log10(sfr1)

logMass2 = np.log10(mass2)
logSfr2 = np.log10(sfr2)
```

```python
# Plotting
plt.figure(figsize=(40,10))
plt.rcParams.update({"font.size": 25})
plt.rcParams.update({'axes.linewidth': 3})

plt.subplot(1, 4, 1)
plt.hist(logMass1, bins=30, color="royalblue", histtype="step", label="$\\rm REFL0100N1504$", linewidth=5)
plt.hist(logMass2, bins=30, color="orange", histtype="step",label="$\\rm NOAGNL0050N0752$", linewidth=5)
plt.xlabel("$\\rm \log M_* \: [M_\odot]$")
plt.ylabel("Number of Galaxies")
plt.xticks([10, 11, 12], [10, 11, 12])
plt.legend()

plt.subplot(1, 4, 2)
plt.hist(logSfr1[logSfr1>-10], bins=30, color="royalblue", histtype="step", linewidth=5)
plt.hist(logSfr2[logSfr2>-10], bins=30, color="orange", histtype="step", linewidth=5)
plt.xlabel("$\\rm \log SFR \: [M_\odot yr^{-1}]$")
plt.ylabel("Number of Galaxies")

plt.subplot(1, 4, 3)
sns.kdeplot(x=logMass1, y=logSfr1, levels=9, color="royalblue", linewidth=2, label="$\\rm REFL0100N1504$", fill=True)
plt.xlabel("$\\rm \log M_* \: [M_\odot]$")
plt.ylabel("$\\rm \log SFR \: [M_\odot yr^{-1}]$")
plt.xlim((8.9, 12.1))
plt.ylim((-3.1, 2.1))
plt.xticks([9, 10, 11, 12], [9, 10, 11, 12])

plt.subplot(1, 4, 4)
sns.kdeplot(x=logMass2, y=logSfr2, levels=9, color="orange", linewidth=2, label="$\\rm NOAGNL0050N0752$", fill=True)
plt.xlabel("$\\rm \log M_* \: [M_\odot]$")
plt.ylabel("$\\rm \log SFR \: [M_\odot yr^{-1}]$")
plt.xlim((8.9, 12.1))
plt.ylim((-3.1, 2.1))
plt.xticks([9, 10, 11, 12], [9, 10, 11, 12])
plt.savefig("SampleDistributions.png", bbox_inches="tight")
```
