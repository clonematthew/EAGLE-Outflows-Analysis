```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing my functions
from dataAnalysisFunctions import *
from outflowProperties import *
```

```python
# Loading in the galaxy data
galaxyData = pd.read_csv("galaxyDataSave2.csv", delimiter=",")

# Removing null selections
galaxyData = removeNulls(galaxyData)

# Only choosing galaxies with an outflow rate (only outflowing/inflowing galaxies)
galaxyData = galaxyData[galaxyData["Mass Outflow Rate"] != 0]

# Selecting only one inclination
galaxyDataCut = galaxyData[galaxyData["Theta"] == "0.0"]
galaxyData = galaxyDataCut[galaxyDataCut["Phi"] == "0.0"]
```

```python
from outflowProperties import *

galaxyData = galaxyOutflowProperties("ANGLE", "0.0", "0.0", 2)
```

```python
#galaxyData = pd.read_csv("galaxyDataSave3.csv", delimiter=",")

mBins, oV, iV, oM, iM, oT, iT, oS, iS = binFlowData(galaxyData, 10)
```

```python
# Logging the values
oMl = np.log10(oM)
iMl = -1 * np.array(iM)
iMl = np.log10(iMl)
```

```python
plt.figure(figsize=(13,10))
plt.rcParams.update({"font.size": 18})
plt.plot(mBins[:-1], oMl, "r", label="Outflow")
plt.plot(mBins[:-1], oMl, "ro", markersize=5)
plt.plot(mBins[:-1], iMl, "b", label="Inflow")
plt.plot(mBins[:-1], iMl, "bo", markersize=5)
plt.xlabel("$\\rm \log M_* \:[M_\odot]$")
plt.ylabel("$\\rm \log \dot{M}_{out/in}\: [M_\odot yr^{-1}]$")
plt.xticks([10, 11], [10, 11])
plt.legend()
```

```python
plt.figure(figsize=(13,10))
plt.rcParams.update({"font.size": 18})
plt.plot(mBins[:-1], oV, "r", label="Outflow")
plt.plot(mBins[:-1], oV, "ro", markersize=5)
plt.plot(mBins[:-1], iV, "b", label="Inflow")
plt.plot(mBins[:-1], iV, "bo", markersize=5)
plt.xlabel("$\\rm \log M_* \:[M_\odot]$")
plt.ylabel("$\\rm v_{out/in} \:[kms^{-1}]$")
plt.xticks([10, 11], [10, 11])
plt.legend()
```
